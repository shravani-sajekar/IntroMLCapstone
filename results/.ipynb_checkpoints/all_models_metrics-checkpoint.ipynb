# Model Comparison and Evaluation

This notebook compares the performance of all models, including classical ML models (Linear Regression, KNN, Neural Network) and literature-based models (Paper 1 XGBoost, Paper 2 RF, Paper 2 XGBoost). 

We compute standard regression metrics (RMSE, MSE, MAE, R²), generate visualizations for comparison, and identify the best-performing model.

## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Model Predictions

Load validation predictions from all models and the actual target values saved during preprocessing. 
This allows us to compute metrics and perform side-by-side comparisons.

```python
# Classic ML models

y_pred_linreg = np.load("../data/lr_val_predictions.npy")
y_pred_knn = np.load("../data/knn_val_predictions.npy")
y_pred_nn = np.load("../data/nn_val_predictions.npy")

# Literature-based models

y_pred_xgb_paper1 = np.load("../data/paper1_xgb_val_predictions.npy")
y_pred_rf_paper2 = np.load("../data/paper2_rf_val_predictions.npy")
y_pred_xgb_paper2 = np.load("../data/paper2_xgb_val_predictions.npy")

# Load actual values

y_val = np.load("../data/y_val.npy") 
```

## Define Metrics Function

Create a helper function to compute RMSE, MSE, MAE, and R² for each model.


```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def compute_metrics(y_true, y_pred):
    rmse = mean_squared_error(y_true, y_pred, squared=False)
    mse  = mean_squared_error(y_true, y_pred)
    mae  = mean_absolute_error(y_true, y_pred)
    r2   = r2_score(y_true, y_pred)
    return round(rmse, 2), round(mse, 2), round(mae, 2), round(r2, 4)
```

## Compute Metrics for All Models

Store metrics in a dictionary for easy conversion to a DataFrame and sorting by R² score.


```python
models = {
    "Linear Regression": y_pred_linreg,
    "KNN Regressor": y_pred_knn,
    "Neural Network": y_pred_nn,
    "Paper 1 XGBoost": y_pred_xgb_paper1,
    "Paper 2 RF": y_pred_rf_paper2,
    "Paper 2 XGBoost": y_pred_xgb_paper2
}

metrics_dict = {}

for name, preds in models.items():
    rmse, mse, mae, r2 = compute_metrics(y_val, preds)
    metrics_dict[name] = {"RMSE": rmse, "MSE": mse, "MAE": mae, "R²": r2}
```

```python
# Convert to DataFrame and sort by R²

metrics_df = pd.DataFrame(metrics_dict).T.sort_values(by="R²", ascending=False)
metrics_df
```

## RMSE Comparison Across Models

Lower RMSE indicates better predictive performance. Visualize the results for all models.


```python
plt.figure(figsize=(10,6))
sns.barplot(x=metrics_df.index, y=metrics_df["RMSE"], palette="viridis")
plt.ylabel("RMSE")
plt.title("RMSE Comparison Across Models")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("../results/rmse_comparison.png", dpi=300)
plt.show()
```

## MAE Comparison Across Models

Lower MAE also indicates better prediction accuracy, complementing RMSE.


```python
plt.figure(figsize=(10,6))
sns.barplot(x=metrics_df.index, y=metrics_df["MAE"], palette="magma")
plt.ylabel("MAE")
plt.title("MAE Comparison Across Models")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("../results/mae_comparison.png", dpi=300)
plt.show()
```

## R² Score Comparison Across Models

R² indicates the proportion of variance explained by the model. Higher is better.

```python
plt.figure(figsize=(10,6))
sns.barplot(x=metrics_df.index, y=metrics_df["R²"], palette="coolwarm")
plt.ylabel("R² Score")
plt.title("R² Comparison Across Models")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("../results/r2_comparison.png", dpi=300)
plt.show()
```

## Best Model: Actual vs Predicted & Residuals

Visualize the predictions of the best-performing model (highest R²).


```python
best_model_name = metrics_df.index[0]

best_preds = models[best_model_name]
print("BEST MODEL : " + best_model_name)
```

```python
# Actual vs Predicted

plt.figure(figsize=(7,7))
plt.scatter(y_val, best_preds, alpha=0.5)
plt.xlabel("Actual SalePrice")
plt.ylabel("Predicted SalePrice")
plt.title(f"{best_model_name}: Actual vs Predicted")
plt.plot([y_val.min(), y_val.max()], [y_val.min(), y_val.max()], 'r--')
plt.tight_layout()
plt.savefig("../results/best_model_actual_vs_pred.png", dpi=300)
plt.show()
```

```python
# Residual plot

residuals = y_val - best_preds
plt.figure(figsize=(7,6))
plt.scatter(best_preds, residuals, alpha=0.5)
plt.axhline(0, color='black', linestyle='--')
plt.xlabel("Predicted SalePrice")
plt.ylabel("Residuals")
plt.title(f"{best_model_name} Residuals")
plt.tight_layout()
plt.savefig("../results/best_model_residuals.png", dpi=300)
plt.show()
```

## Save Metrics

Save metrics as CSV for record-keeping and as an image for visual presentation.

```python
# Save CSV

metrics_df.to_csv("../results/model_comparison_metrics.csv")
```

```python
# Save metrics table as image

def save_df_as_image(df, path="../results/metrics_table.png"):
    # Calculate dynamic size
    rows, cols = df.shape
    fig_width = 5
    fig_height = max(1, rows * 0.5)

    fig, ax = plt.subplots(figsize=(fig_width, fig_height))
    ax.axis("off")

    table = ax.table(
        cellText=df.values,
        colLabels=df.columns,
        loc="center",
        cellLoc="center"
    )

    # Styling
    table.auto_set_font_size(False)
    table.set_fontsize(5)
    table.scale(1, 1.4)

    # Save image
    plt.savefig(path, dpi=300, bbox_inches="tight")
    plt.close()

# Save your metrics df
save_df_as_image(metrics_df)
```
